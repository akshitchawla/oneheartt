```r
food_data  <- read.csv(file = "datasets/food_2.csv")
```

```r
completeFun <- function(data, desiredCols) {
  completeVec <- complete.cases(data[, desiredCols])
  return(data[completeVec, ])
}
```

```r
new_food = completeFun(food_data, "proteins_100g")
```

```r
dim(new_food)
```

```r
new_food = new_food[,c('product_name','proteins_100g')]
```

```r
write.csv(new_food, file = "datasets/food_cals.csv", row.names = F)
```
